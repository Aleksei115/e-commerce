```python
import csv
import random

# Defining categories and their respective brands
categories = {
    "Videojuegos": ["Sony", "Microsoft", "Nintendo", "Ubisoft", "EA"],
    "Telefonía": ["Apple", "Samsung", "Huawei", "Xiaomi", "Nokia"],
    "Suplementos alimenticios": ["Optimum Nutrition", "MuscleTech", "Dymatize", "BSN", "GNC"],
    "Ropa": ["Nike", "Adidas", "Puma", "Under Armour", "H&M"],
    "Skin Care": ["L'Oréal", "Neutrogena", "Olay", "Clinique", "Nivea"],
    "Calzado": ["Nike", "Adidas", "Reebok", "Puma", "Skechers"],
    "Computadoras": ["Dell", "HP", "Apple", "Asus", "Acer"]
}

# Sample products for each category
products = {
    "Videojuegos": ["Console", "Game", "Controller", "Headset", "VR Headset"],
    "Telefonía": ["Smartphone", "Tablet", "Smartwatch", "Charger", "Earbuds"],
    "Suplementos alimenticios": ["Protein Powder", "BCAA", "Pre-Workout", "Creatine", "Vitamins"],
    "Ropa": ["T-Shirt", "Jacket", "Pants", "Shorts", "Sweater"],
    "Skin Care": ["Moisturizer", "Cleanser", "Toner", "Sunscreen", "Serum"],
    "Calzado": ["Running Shoes", "Sneakers", "Boots", "Sandals", "Slippers"],
    "Computadoras": ["Laptop", "Desktop", "Monitor", "Keyboard", "Mouse"]
}

# Sample countries for product manufacturing
countries = ["USA", "China", "Germany", "Japan", "South Korea", "India"]

# Function to generate random products
def generate_products():
    rows = []
    index = 0
    for category, brands in categories.items():
        for brand in brands:
            num_products = random.randint(8, 15)
            for _ in range(num_products):
                prod_index = 'PROD' + str(1000+index)
                index += 1
                product_name = f"{brand} {random.choice(products[category])}"
                price = round(random.uniform(10.0, 1000.0), 2)
                country = random.choice(countries)
                rating = round(random.uniform(1.0, 5.0), 2)
                cantidad = random.randint(100, 500)
                rows.append([prod_index,product_name, price, brand, 
                             country, category, rating
                             , cantidad
                             ])
    return rows

# Generating product data
product_data = generate_products()

# Column headers
headers = ["id_producto","nombre", "precio", "marca", "pais_fabricacion", "categoria", "puntuacion_media","cantidad"]

# Writing to CSV
csv_file_path = "ecommerce_products.csv"
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(headers)
    writer.writerows(product_data)

csv_file_path
```
